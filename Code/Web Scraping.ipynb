```python
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://www.worldometers.info/world-population/population-by-country/"
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.text)
```

```python
data = soup.find_all("table")[0]
```

```python
df_population = pd.read_html(str(data))[0]
df_population.head()
```

```python
print(df_population.columns)

subset_density = df_population['Density  (P/Km²)']
subset_country = df_population['Country (or dependency)']
```

```python
plt.pie(subset_density, labels=subset_country,radius= 2.0)
```
