<h1 style="color:purple;">Datos Categóricos a Visualizar</h1> 

<p>los datos utilizados son obtenidos del dataset PokemonStats.csv se utiliza el dataset por contener una convinacion de datos 
categoricos y datos numericos</p>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_style('whitegrid')
sns.set_palette('mako')

pkm = pd.read_csv("C:\CODE\Python\Py_Mineria\Datasets\PokemonStats.csv", sep = ",")

pkm.head(10)
```

```python
sns.color_palette('mako')
```

<br><br>
<h1 style="color:darkblue;"> Gráficas categóricas de Dispersión</h1>
<h2 style="color:indianred;"> Gráfica stripplot </h2>

```python
sns.catplot(x="Type1", y="Total", data=pkm, palette= 'mako', orient= "v")
plt.show()
```

```python
sns.catplot(x="Type2", y="Total", data=pkm, jitter=False, palette= 'mako')
plt.show()
```

```python
sns.catplot(x="Type1", y="Total", data=pkm,
            kind="strip", hue="Type2", palette= 'mako')
plt.show()
```

```python
sns.catplot(x="Type1", y="Total", data=pkm,
            kind="strip", hue="Attack", col="Type2")
plt.show()
```

```python
sns.catplot(x="Type1", y="Total", data=pkm,
            kind="strip", hue="Type2", s=7, col="Type2")
plt.show()
```

<h2 style="color:indianred;"> Gráfica swarmplot </h2>

```python
sns.catplot(x="Type1", y="Total", data=pkm,
            kind="swarm", hue="Type2", s=7, col="Type2")
plt.show()
```

<br><br>
<h1 style="color:darkblue;"> Gráficas categóricas de Distribución</h1>
<h2 style="color:indianred;"> Gráfica boxplot</h2>

```python
sns.catplot(x="Type1", y="Total", data=pkm,
            kind="box")
plt.show()
```

<h2 style="color:indianred;"> Gráfica boxenplot</h2>

```python
sns.catplot(x="Type1", y="Total", data=pkm, 
            kind="boxen")
plt.show()
```

```python
subset = pkm.loc[pkm['Type1'] == 'Ghost', ['Name', 'Total','Type1', 'Type2']]
```

<h2 style="color:indianred;"> Gráfica violinplot</h2>

```python
sns.catplot(x="Type1", y="Total", data=subset,
            kind="violin")

sns.catplot(x="Type1", y="Total", data=subset,
            kind="violin", hue="Type2")


plt.show()
```

<br><br>
<h1 style="color:darkblue;"> Gráficas categóricas de Estimación</h1>
<h2 style="color:indianred;"> Gráfica pointplot</h2>

```python
sns.catplot(x="Type2", y="Total", data=subset, 
            kind="point")
plt.show()
```

<h2 style="color:indianred;"> Gráfica barplot</h2>

```python
sns.catplot(x="Type2", y="Total", data=subset,
            kind="bar", palette= 'mako')
plt.show()
```

<h2 style="color:indianred;"> Gráfica countplot </h2>

```python
sns.catplot(x="Type1", data=subset,
            kind="count", hue="Type2", palette= 'mako')
plt.show()
```
