```python
import pandas as pd
import matplotlib.pyplot as plt

pokemon = pd.read_csv("C:\CODE\Python\Py_Mineria\Datasets\PokemonStats.csv")

Grass = pokemon[pokemon["Type1"]=='Grass']
Fire = pokemon[pokemon["Type1"]=='Fire']
Water = pokemon[pokemon["Type1"]=='Water']


plt.figure(figsize=(6, 6))
plt.xlabel('Attack', fontsize = 20.0)
plt.ylabel('Defense', fontsize = 20.0)
plt.scatter(Grass["Attack"], Grass["Defense"],
            label="Grass", marker="*", c="chartreuse", s=200)
plt.scatter(Fire["Attack"], Fire["Defense"],
            label="Fire", marker="*", c="red", s=200)
plt.scatter(Water["Attack"], Water["Defense"],
            label="Water", marker="*", c="cyan", s=200)

plt.legend(bbox_to_anchor=(1, 0.15))
plt.show()
```

```python
x= pokemon[["Attack", "Defense"]]

y= pokemon["Type1"]
```

```python
from sklearn import tree

arbol_decision = tree.DecisionTreeClassifier(criterion="entropy")

arbol = arbol_decision.fit(x, y)

print(tree.export_text(arbol, feature_names=["Attack", "Defense"]))
```

```python
plt.figure(figsize=(12, 6))
tree.plot_tree(arbol,
              feature_names=["Attack", "Defense"])
plt.show()
```
